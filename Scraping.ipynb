```python
import requests
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
data = []
```

```python
base_url = 'https://catalog.data.gov/'
url = base_url + 'dataset?q=&sort=views_recent+desc&page=1'
driver = webdriver.Chrome()
```

```python
for i in range(0, 300):  # Loop untuk 30 halaman
    print(f"Scraping Halaman {i+1}")

    # Selenium membuka halaman untuk memastikan views muncul
    driver.get(url)
    time.sleep(5)  # Tunggu agar JavaScript selesai menampilkan angka views

    # Ambil source code halaman setelah dimuat oleh Selenium
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    data_gov = soup.find_all('div', class_='dataset-content')

    for datagov in data_gov:
        # Ambil judul dataset
        title_element = datagov.find('h3', class_="dataset-heading")
        title = title_element.find('a').text.strip() if title_element and title_element.find('a') else "Tidak ada judul"

        # Ambil jumlah views dengan Selenium
        views_element = datagov.find('span', class_='recent-views recent-views-datagov')
        view_count = views_element['title'] if views_element else "Tidak ditemukan"

        # Ambil catatan dataset
        notes_element = datagov.find('div', class_='notes')
        Notes = notes_element.text.strip() if notes_element else "Tidak ada catatan"

        # Simpan data
        data.append({
            'Title': title,
            'Views': view_count,
            'Notes': Notes
        })

    # Cek tombol "next page" untuk navigasi ke halaman berikutnya
    next_page = soup.find('a', class_='page-link', string='»')
    if next_page and 'href' in next_page.attrs:
        url = base_url + next_page['href']
    else:
        break

    time.sleep(3) 

# Tutup Selenium
driver.quit()

# Simpan data ke CSV
df = pd.DataFrame(data, columns=['Title', 'Views', 'Notes'])
df.to_csv('scrapingdatagov.csv', index=False)

print(f"Total data yang dikumpulkan: {len(data)}")
print(data)
```
