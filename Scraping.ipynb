```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
data = [] # Digunakan untuk menyimpan data baru
```

```python
url = 'https://catalog.data.gov/dataset?q=&sort=views_recent+desc' #Baris ini mendeklarasikan variabel bernama url dan menyimpan alamat URL situs web yang ingin diakses.
Response = requests.get(url) # Baris ini memanggil fungsi get() dari library requests untuk melakukan permintaan HTTP GET ke URL yang telah ditentukan dan hasilnya disimpan dalam variabel bernama Response yang akan berisi semua informasi respons dari server, termasuk status kode, header, dan konten halaman web.
```

```python
for i in range(0, 300):
    soup = BeautifulSoup(Response.text, 'html.parser')
    data_gov = soup.find_all('div', class_='dataset-content')

    for datagov in data_gov:

        title_element = datagov.find('h3', class_="dataset-heading")
        if title_element and title_element.find('a'):
          title = title_element.find('a').text.strip()
        else:
          title = "Tidak ada judul"

        Notes = datagov.find('div', class_='notes').text
        data.append({
            'Title': title,
            'Notes': Notes
            })

    time.sleep(1)
    href_value = soup.find('a', class_='page-link')['href']
    time.sleep(3)

print(f"Total data yang dikumpulkan: {len(data)}")
print(data)

df = pd.DataFrame(data, columns=['Title', 'Notes'])
df.to_csv('scraping.csv', index=False)
```
